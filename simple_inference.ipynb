```python
from utils.model import SiameseModel
from torchvision import transforms
from PIL import Image
import matplotlib.pyplot as plt
import torch.nn.functional as F
import torch
import cv2
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
device
```

```python
# Instantiate the model
model = SiameseModel()
```

```python
# Load the saved weights
model.load_state_dict(torch.load("model_weights_rmsprop.pth"))
```

```python
# Set the model to evaluation mode
model.eval()
```

```python
transform = transforms.Compose([
    transforms.Resize((256,256)),
    transforms.ToTensor()
])
```

```python
img1 = Image.open("Input_images/captured.jpg")
img2 = Image.open("Owner_identification/Rajat/IMG-20241201-WA0012_face0.jpg")
```

```python
img1, img2 = transform(img1), transform(img2)
```

```python
img1 = img1.unsqueeze(0)
img2 = img2.unsqueeze(0)
```

```python
img2.shape
```

```python
emb1, emb2 = model(img1, img2)
```

```python
distance = F.pairwise_distance(emb1, emb2)
print(distance.item())
```
